### An치lisis Exploratorio de Datos (EDA)

#### Contexto del Dataset
**Nombre:** `taxi_trip_data.csv`  
**Tama침o:** 18 columnas y 22,699 filas  
El an치lisis se centra en tres variables clave del proyecto Automatidata:
- **trip_duration:** Duraci칩n del viaje en minutos.
- **trip_distance:** Distancia del viaje en millas.
- **total_amount:** Monto total cobrado, incluyendo tarifa, extras, impuestos y propinas (excepto propinas en efectivo).

#### 1. Visualizaci칩n de Distribuciones
#### Objetivo:
Analizar la forma, dispersi칩n y valores at칤picos de las variables clave, con consideraciones de accesibilidad para no expertos y personas con discapacidad visual.

##### Gr치ficos Utilizados:
1. **Histograma con KDE (Kernel Density Estimation):**
   - Proporciona una vista de la forma de la distribuci칩n.
   - Identifica concentraciones de datos y la presencia de m칰ltiples picos.

2. **Diagrama de caja (Boxplot):**
   - Destaca valores at칤picos y distribuciones fuera del rango intercuartil.

3. **Resumen estad칤stico:**
   - Calcula medidas centrales y dispersi칩n (media, mediana, desviaci칩n est치ndar).


```python
# Importar librer칤as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo accesible
plt.style.use('seaborn-v0_8-dark')  # Alto contraste
sns.set_context("talk")  # Fuentes grandes

# Cargar el dataset
file_path = "C:/Users/victo/OneDrive/Data_Science/Python/taxi_trip_data_actualizado.csv"
df = pd.read_csv(file_path)

# Convertir fechas a datetime y calcular trip_duration
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
df['trip_duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60

# Filtrar inconsistencias b치sicas
df = df[(df['trip_duration'] >= 0) & (df['trip_distance'] >= 0) & (df['total_amount'] >= 0)]

# Lista de variables a visualizar
variables = ['trip_duration', 'trip_distance', 'total_amount']
labels = {
    'trip_duration': 'Duraci칩n del viaje (minutos)',
    'trip_distance': 'Distancia del viaje (millas)',
    'total_amount': 'Monto total ($)'
}
limits = {
    'trip_duration': (0, 100),  # Limitar para visualizaci칩n
    'trip_distance': (0, 50),
    'total_amount': (0, 200)
}

# Crear gr치ficos para cada variable
for var in variables:
    # Histograma con KDE
    plt.figure(figsize=(12, 6))
    sns.histplot(df[var], bins=50, kde=True, color='coral')
    plt.title(f'Distribuci칩n de {labels[var]}', fontsize=16)
    plt.xlabel(labels[var], fontsize=14)
    plt.ylabel('Frecuencia', fontsize=14)
    plt.xlim(limits[var])
    # A침adir texto para accesibilidad
    plt.text(0.8 * limits[var][1], plt.gca().get_ylim()[1] * 0.9, 
             f"Media: {df[var].mean():.1f}\nMediana: {df[var].median():.1f}", 
             fontsize=12, bbox=dict(facecolor='white', alpha=0.8))
    plt.grid(True)
    plt.show()
    
    # Diagrama de caja
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=df[var], color='lightblue')
    plt.title(f'Boxplot de {labels[var]}', fontsize=16)
    plt.xlabel(labels[var], fontsize=14)
    plt.xlim(limits[var])
    plt.grid(True)
    plt.show()

# Resumen estad칤stico
print("Resumen estad칤stico de las variables:")
display(df[variables].describe())
```

### Evaluaci칩n de Variables Temporales. Estas dos variables seran parte del futuro modelo de Regresion del Proyecto

#### Objetivo:
Analizar la hora de inicio y finalizaci칩n del viaje, derivadas de las columnas temporales del dataset, para capturar patrones relevantes en la demanda y duraci칩n de los viajes. Estas variables pueden ser 칰tiles para enriquecer futuros modelos de regresi칩n.

#### Variables:
1. **pickup_hour:** Hora de inicio del viaje, extra칤da de la columna `tpep_pickup_datetime`.
   - **Justificaci칩n:** Ayuda a identificar per칤odos de alta o baja demanda, relacionados con patrones temporales en los viajes (horas pico, nocturnas, etc.).

2. **dropoff_hour:** Hora de finalizaci칩n del viaje, extra칤da de la columna `tpep_dropoff_datetime`.
   - **Justificaci칩n:** Permite evaluar la duraci칩n de los viajes en relaci칩n con el horario de finalizaci칩n, lo que puede ser 칰til para analizar la disponibilidad y flujo de taxis en diferentes momentos del d칤a.

#### Pasos para Generar las Variables:
1. Convertir las columnas `tpep_pickup_datetime` y `tpep_dropoff_datetime` a formato de fecha y hora (`datetime`).
2. Extraer las horas de inicio y finalizaci칩n como nuevas variables:
   - **pickup_hour**: `tpep_pickup_datetime.dt.hour`
   - **dropoff_hour**: `tpep_dropoff_datetime.dt.hour`

#### Herramientas Utilizadas:
- **Pandas:** Para manipulaci칩n de datos y extracci칩n de las variables temporales.
- **Python:** Lenguaje base para construir las transformaciones necesarias.

#### Resultados Esperados:
- Visualizaci칩n clara de las distribuciones horarias (`pickup_hour` y `dropoff_hour`) para identificar tendencias y patrones.
- Generaci칩n de insights que enriquezcan futuros modelos predictivos de regresi칩n.


```python
# Importar librer칤as
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo accesible
plt.style.use('seaborn-v0_8-dark')  # Alto contraste
sns.set_context("talk")  # Fuentes grandes

# Cargar el dataset original
file_path = "C:/Users/victo/OneDrive/Data_Science/Python/taxi_trip_data.csv"
df = pd.read_csv(file_path)

# Crear columnas nuevas
# 1. trip_duration (en minutos)
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
df['trip_duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60

# 2. Month (formato YYYY-MM)
df['Month'] = df['tpep_pickup_datetime'].dt.to_period('M')

# 3. Day (d칤a de la semana, ej. Monday)
df['Day'] = df['tpep_pickup_datetime'].dt.day_name()

# Verificar nuevas columnas
print("Primeras filas con nuevas columnas:")
display(df[['tpep_pickup_datetime', 'trip_duration', 'Month', 'Day']].head())

# Filtrar inconsistencias (duraciones negativas)
df = df[df['trip_duration'] >= 0]
print("\nFilas tras filtrar duraciones negativas:", len(df))

# Recuento total de viajes por mes
trips_by_month = df.groupby('Month').size().reset_index(name='trip_count')
print("\nRecuento de viajes por mes:")
display(trips_by_month)

# Gr치fico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='trip_count', data=trips_by_month, color='teal')
plt.title('N칰mero de viajes en taxi por mes', fontsize=16)
plt.xlabel('Mes', fontsize=14)
plt.ylabel('N칰mero de viajes', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
# A침adir texto para accesibilidad
plt.text(0.05, 0.95, f"Total viajes: {trips_by_month['trip_count'].sum()}", 
         transform=plt.gca().transAxes, fontsize=12, bbox=dict(facecolor='white', alpha=0.8))
plt.tight_layout()
plt.show()

# Guardar el dataset transformado con un nuevo nombre
new_file_path = "C:/Users/victo/OneDrive/Data_Science/Python/taxi_trip_data_transformed.csv"
df.to_csv(new_file_path, index=False)
print(f"\nDataset transformado guardado como: {new_file_path}")
```

```python
# Identificar outliers con IQR para trip_duration
Q1 = df['trip_duration'].quantile(0.25)
Q3 = df['trip_duration'].quantile(0.75)
IQR = Q3 - Q1
outliers_duration = df[(df['trip_duration'] < (Q1 - 1.5 * IQR)) | (df['trip_duration'] > (Q3 + 1.5 * IQR))]
print("N칰mero de outliers en duraci칩n:", len(outliers_duration))

# Media vs. Mediana
print("\nMedia de trip_duration:", df['trip_duration'].mean())
print("Mediana de trip_duration:", df['trip_duration'].median())

# Ejemplo de outliers
print("\nEjemplo de viajes con duraci칩n extrema:")
display(outliers_duration[['trip_duration', 'trip_distance', 'total_amount']].head())
```
