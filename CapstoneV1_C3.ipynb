### Análisis Exploratorio de Datos (EDA)

#### Contexto del Dataset
**Nombre:** `taxi_trip_data.csv`  
**Tamaño:** 18 columnas y 22,699 filas  
El análisis se centra en tres variables clave del proyecto Automatidata:
- **trip_duration:** Duración del viaje en minutos.
- **trip_distance:** Distancia del viaje en millas.
- **total_amount:** Monto total cobrado, incluyendo tarifa, extras, impuestos y propinas (excepto propinas en efectivo).

#### 1. Visualización de Distribuciones
#### Objetivo:
Analizar la forma, dispersión y valores atípicos de las variables clave, con consideraciones de accesibilidad para no expertos y personas con discapacidad visual.

##### Gráficos Utilizados:
1. **Histograma con KDE (Kernel Density Estimation):**
   - Proporciona una vista de la forma de la distribución.
   - Identifica concentraciones de datos y la presencia de múltiples picos.

2. **Diagrama de caja (Boxplot):**
   - Destaca valores atípicos y distribuciones fuera del rango intercuartil.

3. **Resumen estadístico:**
   - Calcula medidas centrales y dispersión (media, mediana, desviación estándar).


```python
# Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar estilo accesible
plt.style.use('seaborn-v0_8-dark')  # Alto contraste
sns.set_context("talk")  # Fuentes grandes

# Cargar el dataset
file_path = "C:/Users/victo/OneDrive/Data_Science/Python/taxi_trip_data_actualizado.csv"
df = pd.read_csv(file_path)

# Convertir fechas a datetime y calcular trip_duration
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
df['trip_duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60

# Filtrar inconsistencias básicas
df = df[(df['trip_duration'] >= 0) & (df['trip_distance'] >= 0) & (df['total_amount'] >= 0)]

# Lista de variables a visualizar
variables = ['trip_duration', 'trip_distance', 'total_amount']
labels = {
    'trip_duration': 'Duración del viaje (minutos)',
    'trip_distance': 'Distancia del viaje (millas)',
    'total_amount': 'Monto total ($)'
}
limits = {
    'trip_duration': (0, 100),  # Limitar para visualización
    'trip_distance': (0, 50),
    'total_amount': (0, 200)
}

# Crear gráficos para cada variable
for var in variables:
    # Histograma con KDE
    plt.figure(figsize=(12, 6))
    sns.histplot(df[var], bins=50, kde=True, color='coral')
    plt.title(f'Distribución de {labels[var]}', fontsize=16)
    plt.xlabel(labels[var], fontsize=14)
    plt.ylabel('Frecuencia', fontsize=14)
    plt.xlim(limits[var])
    # Añadir texto para accesibilidad
    plt.text(0.8 * limits[var][1], plt.gca().get_ylim()[1] * 0.9, 
             f"Media: {df[var].mean():.1f}\nMediana: {df[var].median():.1f}", 
             fontsize=12, bbox=dict(facecolor='white', alpha=0.8))
    plt.grid(True)
    plt.show()
    
    # Diagrama de caja
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=df[var], color='lightblue')
    plt.title(f'Boxplot de {labels[var]}', fontsize=16)
    plt.xlabel(labels[var], fontsize=14)
    plt.xlim(limits[var])
    plt.grid(True)
    plt.show()

# Resumen estadístico
print("Resumen estadístico de las variables:")
display(df[variables].describe())
```

### 6. Identificación de Anomalías o Patrones Interesantes

#### Objetivo:
Detectar outliers en el dataset utilizando métodos sugeridos, como diagramas de caja, comparación entre medidas centrales (media/mediana) e histogramas. Decidir cómo manejar estos valores para garantizar que el análisis y modelado sean precisos y consistentes.

#### Métodos Utilizados:
1. **Diagrama de Caja (Boxplot):**
   - Identificar valores fuera del rango intercuartil.
   - Visualización clara de la distribución y posibles outliers extremos.

2. **Comparación entre Media y Mediana:**
   - Analizar discrepancias entre la media y la mediana para identificar columnas con distribuciones sesgadas.
   - Detectar valores atípicos que influyen significativamente en la media.

3. **Histogramas:**
   - Detectar patrones inesperados, como distribuciones bimodales o concentraciones fuera de lo común.
   - Facilitar la visualización de la densidad de datos en diferentes rangos.

#### Pasos:
1. **Generar visualizaciones para las variables clave**:
   - `trip_duration`
   - `trip_distance`
   - `total_amount`
   
   Utilizando diagramas de caja e histogramas para cada variable.

2. **Calcular medidas centrales**:
   - Comparar la media y la mediana para determinar si los datos tienen distribuciones sesgadas.

3. **Decidir cómo manejar los outliers**:
   - **Opción 1:** Imputación (si los valores son relevantes pero erróneos).
   - **Opción 2:** Eliminación de outliers extremos (si no aportan al análisis).


```python
# Identificar outliers con IQR para trip_duration
Q1 = df['trip_duration'].quantile(0.25)
Q3 = df['trip_duration'].quantile(0.75)
IQR = Q3 - Q1
outliers_duration = df[(df['trip_duration'] < (Q1 - 1.5 * IQR)) | (df['trip_duration'] > (Q3 + 1.5 * IQR))]
print("Número de outliers en duración:", len(outliers_duration))

# Media vs. Mediana
print("\nMedia de trip_duration:", df['trip_duration'].mean())
print("Mediana de trip_duration:", df['trip_duration'].median())

# Ejemplo de outliers
print("\nEjemplo de viajes con duración extrema:")
display(outliers_duration[['trip_duration', 'trip_distance', 'total_amount']].head())
```
