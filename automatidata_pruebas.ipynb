```python
import pandas as pd
import numpy as np

# Aseg√∫rate de escribir 'read_csv' correctamente
df = pd.read_csv("C:/Users/victo/OneDrive/Data_Science/Python/taxi_trip_data_actualizado.csv")

# Imprime el DataFrame
print(df)
print
df.head(10)
```

```python

import pandas as pd

# Asegurarnos de que las columnas est√©n en formato datetime
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

# Calcular duraci√≥n del viaje en minutos
df['trip_duration'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 3600

# Obtener las 10 duraciones m√°s largas
longest_trips = df.nlargest(10, 'trip_duration')[['trip_duration', 'tpep_pickup_datetime', 'tpep_dropoff_datetime']]

# Imprimir resultados con √≠ndice
print("Las 10 duraciones m√°s largas son:")
print(longest_trips)

```

```python

import seaborn as sns  # Aseg√∫rate de importar seaborn
import matplotlib.pyplot as plt  # Para graficar

# Crear un histograma de la columna 'trip_distance'
sns.histplot(df['trip_distance'], bins=30, kde=True)
plt.title("Distribuci√≥n de la distancia del viaje")
plt.xlabel("Distancia (millas)")
plt.show()

```

```python
# identicar viajes largos anomalos
# Umbral para definir viajes largos
long_trips = df[df['trip_distance'] > 20]
print(f"Viajes largos encontrados: {len(long_trips)}")
```

```python
# ingresos totales por tipo de pago
total_income = df.groupby('payment_type')['total_amount'].sum()
print(total_income)
total_income.plot(kind='bar', title='Ingresos totales por tipo de pago')
plt.ylabel('Total ($)')
plt.show()
```

```python
# mapear codigos a descripcciones legibles
rate_code_map = {
    1: "Tarifa est√°ndar",
    2: "JFK",
    3: "Newark",
    4: "Nassau o Westchester",
    5: "Tarifa negociada",
    6: "Viaje en grupo"
}

df['RatecodeID'] = df['RatecodeID'].map(rate_code_map)
print(df[['RatecodeID']].head())
```
